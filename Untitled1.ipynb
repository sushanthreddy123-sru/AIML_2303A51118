part-02: Lab task is to visualize the data set using matplotlib

```python
people = ['prabhas','charan','vijay','lesnar','ntr']
age = [40,35,44,47,37]
weight = [65,60,67,70,58]
height = [155,140,153,160,154]

import matplotlib.pyplot as plt

plt.scatter(weight, height)
plt.title("relationship btw weight and height of people")
plt.ylabel("Height in meters")
plt.xlabel("Weight in kgs")
plt.show()
plt.figure(figsize=(12,4))
plt.title("peoples weight in kgs", fontsize=16,fontstyle='italic')
#main plot function 'bar'
plt.bar(x=people,height=weight, width=0.6,color='orange',edgecolor='k',alpa=0.6)
plt.xlabel("people",fontsize=15)
plt.ylabel("weight(in kgs)",fontsize=15)
plt.show()
```
