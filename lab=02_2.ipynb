```python
graph1 = {
  'A': ['B', 'C', 'G'],
  'B': ['A'],
  'C': ['A' 'D', 'F',],
  'D': ['A', 'C', 'E', 'H'],
  'E': ['D', 'H', 'F'],
  'F': ['C', 'E'],
  'G': ['A', 'D', 'H'],
  'H': ['D', 'E', 'G']

}


def dfs(graph, node, visited):
   if node not in visited:
     visited.append(node)
     for k in graph[node]:
         dfs(graph, k , visited)
   return visited
visited = dfs(graph1, 'B', [])
print(visited)


```
